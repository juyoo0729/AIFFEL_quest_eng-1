```python
import math

def calculator():
    print("=== 간단한 계산기 ===")

    while True:
        try:
            # 첫 번째 정수 입력
            num1_input = input("\n첫 번째 정수를 입력하세요: ").strip()
            num1 = int(num1_input)

            # 연산자 입력
            operator = input("연산자를 입력하세요 (+, -, *, /, **): ").strip()

            # 두 번째 정수 입력
            num2_input = input("두 번째 정수를 입력하세요: ").strip()
            num2 = int(num2_input)

            # 연산 수행
            if operator == '+':
                result = num1 + num2
            elif operator == '-':
                result = num1 - num2
            elif operator == '*':
                result = num1 * num2
            elif operator == '/':
                if num2 == 0:
                    raise ZeroDivisionError
                result = num1 / num2
            elif operator == '**':
                result = math.pow(num1, num2)   # math 모듈 사용
            else:
                raise ValueError(f"지원하지 않는 연산자입니다: {operator}")

            # 결과 출력
            if operator == '/' and result == int(result):
                print(f"결과: {num1} {operator} {num2} = {int(result)}")
            else:
                print(f"결과: {num1} {operator} {num2} = {result}")

        except ValueError as e:
            if "invalid literal for int()" in str(e) or "지원하지 않는 연산자" in str(e):
                if "invalid literal for int()" in str(e):
                    print("잘못된 입력입니다. 정수를 입력해주세요.")
                else:
                    print(f"오류: {e}")
            else:
                print("잘못된 입력입니다. 정수를 입력해주세요.")

        except ZeroDivisionError:
            print("오류: 0으로 나눌 수 없습니다.")

        except Exception as e:
            print(f"알 수 없는 오류가 발생했습니다: {e}")

        # 계속 계산할지 묻기
        while True:
            continue_calc = input("\n계속 계산하시겠습니까? (y/n): ").strip().lower()
            if continue_calc in ['y', 'yes', '예']:
                break
            elif continue_calc in ['n', 'no', '아니오']:
                print("계산기를 종료합니다. 감사합니다!")
                return
            else:
                print("y 또는 n으로 입력해주세요.")

# 프로그램 실행
if __name__ == "__main__":
    calculator()
```
